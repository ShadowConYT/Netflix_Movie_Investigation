````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.read_csv('netflix_data.csv')
netflix.dropna(inplace=True)
netflix.head()
````

### **1. What are the most common genres on Netflix?**

````python
colors = []

for i, row in netflix.iterrows():
    if row['genre'] == 'Dramas':
        colors.append('red')
    elif row['genre'] == 'Horror Movies':
        colors.append('orange')
    elif row['genre'] == 'Action':
        colors.append('yellow')
    elif row['genre'] == 'International TV':
        colors.append('green')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Independent Movies':
        colors.append('purple')
    elif row['genre'] == 'Comedies':
        colors.append('pink')
    elif row['genre'] == 'Sci-Fi':
        colors.append('brown')
    elif row['genre'] == 'International Movies':
        colors.append('gray')
    elif row['genre'] == 'Children':
        colors.append('black')
    elif row['genre'] == 'TV Shows':
        colors.append('red')
    elif row['genre'] == 'Uncategorized':
        colors.append('orange')
    elif row['genre'] == 'Classic Movies':
        colors.append('yellow')
    elif row['genre'] == 'Thrillers':
        colors.append('green')
    elif row['genre'] == 'Stand-Up':
        colors.append('blue')
    elif row['genre'] == 'Anime Features':
        colors.append('purple')
    elif row['genre'] == 'Music':
        colors.append('pink')
    elif row['genre'] == 'Anime Series':
        colors.append('brown')
    elif row['genre'] == 'Kids':
        colors.append('gray')
    elif row['genre'] == 'Docuseries':
        colors.append('black')
    elif row['genre'] == 'Crime TV':
        colors.append('red')
    elif row['genre'] == 'British TV':
        colors.append('orange')
    elif row['genre'] == 'Cult Movies':
        colors.append('yellow')
    elif row['genre'] == 'TV Action':
        colors.append('green')
    elif row['genre'] == 'Romantic TV':
        colors.append('blue')
    elif row['genre'] == 'TV Horror':
        colors.append('purple')
    elif row['genre'] == 'Romantic Movies':
        colors.append('pink')
    elif row['genre'] == 'TV Comedies':
        colors.append('brown')
    elif row['genre'] == 'Classic':
        colors.append('gray')

len(colors)
````

````python
fig = plt.figure(figsize=(12,8))
plt.hist(netflix['genre'], bins=20, color='blue', ec='black')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Distribution of Netflix Titles by Genre')
plt.show()
````

##### So the Most Common Genre on Netflix is **Drames** with 1928 movies, followed by **Comedies** with 1207 movies and **Documentaries** with 644 movies.

### **2. Which countries produce the most content on Netflix?**

````python
netflix_movies = netflix[netflix['type'] == 'Movie']
netflix_series = netflix[netflix['type'] == 'TV Show']

figg = plt.figure(figsize=(12,8))
plt.hist(netflix_movies['country'][:10], color='blue', ec='black')
plt.hist(netflix_series['country'][:10], color='red', ec='black')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Distribution of Netflix Movies by Country')
plt.show()
````
